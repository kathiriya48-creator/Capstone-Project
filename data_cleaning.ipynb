```python
import pandas as pd   # imports pandas to read, clean, and save dataset tables.
```

```python
df = pd.read_csv("/content/energy_dataset.csv")      # loads raw dataset into Python
```

```python
print(df.head())
```

```python
print(df.columns.tolist())         # shows all column names so we can pick the required ones
```

```python
df = df[["time", "total load actual", "price day ahead"]]      # keeps only the 3 columns needed for capstone.
```

```python
df = df.rename(columns={
    "total load actual": "load",
    "price day ahead": "price"
})
```

```python
print(df.head())
```

```python
df["time"] = pd.to_datetime(df["time"], utc=True)         # converts the time column into proper datetime format for time series work.
```

```python
df = df.sort_values("time")         # arranges rows from oldest time to newest time
```

```python
df = df.reset_index(drop=True)     # resets row numbers after sorting.
```

```python
print(df.isnull().sum())       # counts missing values in each column.
```

```python
print("Duplicate timestamps:", df["time"].duplicated().sum())      # checks whether any time value appears more than once.
```

```python
df = df.dropna(subset=["time", "load", "price"])       # removes rows where key variables are missing.
```

```python
df = df.drop_duplicates(subset=["time"])               # removes repeated timestamps to keep one row per hour.
```

```python
print(df.info())
```

```python
print(df.describe())               # gives summary statistics to inspect load and price values.
```

```python
df.to_csv("energy_clean.csv", index=False)          # saves the cleaned final dataset for the next project step
```
